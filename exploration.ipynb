# Ob() Field Simulation Exploration

Use this notebook to experiment with wave parameters and emergent patterns.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-9, 9, 420)
y = np.linspace(-9, 9, 420)
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

t = 0.0
Z = (
    np.sin(1.15*X + 1.55*Y + t)
    + np.sin(1.85*X - 1.25*Y - 0.75*t)
    + 0.9*np.sin(1.15*R - 1.4*t)
    + 0.65*np.sin(4*theta + 0.9*np.sin(t))
    + 0.35*np.sin(0.28*X*Y + 0.5*t)
)
Z = np.tanh(Z)

plt.figure(figsize=(7, 7))
plt.imshow(Z, extent=[-9, 9, -9, 9], origin='lower')
plt.axis('off')
plt.show()
```
